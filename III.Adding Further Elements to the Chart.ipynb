# Adding Further Elements to the Chart:
In order to make a chart more informative, many times it is not enough to represent the data using lines or markers and assign the range of values using two axex.  
In fact, there are many other elements that can be added to a chart in order to enrich it with additional information.  
In this section you will see how to add additional elements to the chart as text labels, a legend, and so on.  

### Adding Text:
You've already seen how you can add the title to the chart by __title()__ function. Tow other textual indications very important to be added to a chart the __axis labels__.   
This is possible through the use of two other specific functions such as __xlabel()__ and __ylabel()__ .  
These functions take as argument a string which will be shown text.

```python
import matplotlib.pyplot as plt
import math
import numpy as np
```

```python
plt.axis([0,5,0,20])
plt.title('My first plot')
plt.xlabel('A plot is more informative by adding axis labels')
plt.ylabel('Square values')
plt.plot([1,2,3,4],[1,4,9,16],'ro')
```

Thanks to the keywords, you can change the characteristics of the text. For example, you can modify the title by changing the font and increasing the size of the characters. Aslo you can modify the color of the axis labels to bring out more the title of the plot

```python
plt.axis([0,5,0,20])
plt.title('My first plot', fontsize=20, fontname='Times New Roman')
plt.xlabel('Counting',fontsize=20, color='blue')
plt.ylabel('Square values',fontsize=20, color='green')
plt.plot([1,2,3,4],[1,4,9,16],'r', linewidth=5)
```

But matplotlib is not limited to this: pyplot allows you to add text to any position within a chart. This feature is preformed by a specific fucntion __text()__.  
$$text(x,y,s,fontdict=None, **kwargs)$$  

The first two arguments are the coordinates of the location where you want to place the text.  
__s__ is the string of text to be added, and __fontdict__ (optional) is the font that you want the text to be represented.  
Finally, you can add the keywords.  
Thus add the label to each point of the plot.Because the first two arguments to the __text()__ function are the coordinates of the graph, you have to use the coorinates of the four points of the plot shifted slightly on the y axis.

```python
plt.axis([0,5,0,20])
plt.title('My first plot',fontsize=20,fontname='Times New Roman')
plt.xlabel('Counting',color='gray')
plt.ylabel('Square values',color='gray')
plt.text(1,1.5,'First')
plt.text(2,4.5,'Second')
plt.text(3,9.5,'Third')
plt.text(4,16.5,'Fourth')
plt.plot([1,2,3,4],[1,4,9,16], 'ro')
```

As you can see now each point of the plot has its label reporting a description.


### LaTex expressions:


Since matplotlib is a graphic library designed to be used in scientific circles, it must be able to exploit the full potential of scentific language, including mathimatical expressions. Matplotlib offers the possibility to integrate LaTex expressions allowing you to insert mathimatical expressions within the chart.  
To do this you can add a LaTex expressions to the text, enclosing it between two ($) characters. The interperter will recongnise them as LaTex expressions and convert them in the corresponding graphic, which can be mathematical expression, a formula, mathematical characters, or just greec letters. Generally you have to precede the string containing LaTex expressions with an 'r', which indicates row text in order to avoid unintended escpae sequence.    

Here, you can also make use of the keywords to further enrich the text to be shown in the plot.  
Therefore, as an example, you can add the formula describing the trend followed by the point of the plot and enclose it in a colored bounding box

```python
plt.axis([0,5,0,20])
plt.title('My first plot', fontsize=20, fontname='Times New Roman')
plt.xlabel('Counting', color='grey')
plt.ylabel('Square values', color='grey')
plt.text(1,1.5,'First')
plt.text(2,4.5,'Second')
plt.text(3,9.5,'Third')
plt.text(4,16.5, 'Fourth')
plt.text(1.1,12,r'$y = x^2$', fontsize=20, bbox={'facecolor':'yellow','alpha':0.2})
plt.plot([1,2,3,4],[1,4,9,16], 'ro')
```

### Adding a Grid:
Another element you can add to a plot is a grid. Often its addition is necessary in order to better understand the position occupied by each point on the chart.  
Adding a grid to chart is very simple operation: just add the __grid()__ function, passing $True$ as argument.  

```python
plt.axis([0,5,0,20])
plt.title('My first plot',fontsize=20,fontname='Times New Roman')
plt.xlabel('Counting',color='gray')
plt.ylabel('Square values',color='gray')
plt.text(1,1.5,'First')
plt.text(2,4.5,'Second')
plt.text(3,9.5,'Third')
plt.text(4,16.5,'Fourth')
plt.text(1.1,12,r'$y = x^2$',fontsize=20,bbox={'facecolor':'yellow','alpha':0.2})
plt.grid(True)
plt.plot([1,2,3,4],[1,4,9,16],'ro')
```

### Adding a Legend:
Another very important component that should be present in any chart is the legend. pyplot also providesa specific function for this type of object:__legend()__.  
Add a lengend to your chart with the  __legend()__ function and string indicating the words with which you want the series to be shown.  
In this example you will assign the 'First series' name to the input data array

```python
plt.axis([0,5,0,20])
plt.title('My first plot',fontsize=20,fontname='Times New Roman')
plt.xlabel('Counting',color='gray')
plt.ylabel('Square values',color='gray')
plt.text(1,1.5,'First')
plt.text(2,4.5,'Second')
plt.text(3,9.5,'Third')
plt.text(4,16.5,'Fourth')
plt.text(1.1,12,r'$y = x^2$',fontsize=20,bbox={'facecolor':'yellow','alpha':0.2})
plt.grid(True)
plt.plot([1,2,3,4],[1,4,9,16],'ro')
plt.legend(['First series'])
```

As you can see the legend is added in the upper-right corner by default. Again if you want to change this behavior you will need to add a few kwargs.  

### Legend Location:
For example, the position occupied by the legend is set by assigning numbers from 0 to 10 th the __loc__ kwargs.  
Each of thses numbers characterized one of the corners of the chart.   
A value of 1 is the default, that is, the upper-right corner.

Before you begin to modify  the code to move the legend, i want to add a small notice. Generally, the legends are used to indicate the definition of series to the reader via a label associated with a color and/or a marker that distinguishes it in the plot. So far, in the examples you have used a single series that was expressed by a single __plot()__ function.  
Now, you have to focus on a more general case in which the same plot shows more series simultaneously.  
Each seriesin the chart will be characterized by a specific color and a specific marker.  
In terms of code, instead, each series will be characterized by a call to the __plot()__ function and the order in which they are defined will correspond to the order of the text labels passed as argument to the __legend()__ function.  

```python
plt.axis([0,5,0,20])
plt.title('My first plot',fontsize=20,fontname='Times New Roman')
plt.xlabel('Counting',color='gray')
plt.ylabel('Square values',color='gray')
plt.text(1,1.5,'First')
plt.text(2,4.5,'Second')
plt.text(3,9.5,'Third')
plt.text(4,16.5,'Fourth')
plt.text(1.1,12,'$y = x^2$',fontsize=20,bbox={'facecolor':'yellow','alpha':0.2})
plt.grid(True)
plt.plot([1,2,3,4],[1,4,9,16],'ro')
plt.plot([1,2,3,4],[0.8,3.5,8,15],'g^')
plt.plot([1,2,3,4],[0.5,2.5,4,12],'b*')
plt.legend(['First series','Second series','Third series'],loc=2)
```
